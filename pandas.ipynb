'read_csv' is the command used to read a comma separated file(csv). Simillarly, in Pandas 'read_excel' is used to read excel files. Pandas reads all the blank values 'NaN'. 

Reading csv files sometimes create an 'Unnnamed: 0' as the first column of the dataframe. So, to avoid this we can specify an argument 'index_col= 0' to read_csv() function, then it reads in the first column as the index.

In the Toypta dataset some values are neither blank nor NaN, they may have some special characters like "??", etc. These values are replaced with NaN using an arguement 'na_values = ['list of unwanted values']' to read_csv() function, now pandas reads these unknown values as NaN.

```python
data = pd.read_csv('Toyota.csv',index_col=0,na_values=["??","????"])

data
```

It is possible that we sometimes copy of data may be needed because it is possible that we don't want to change the original data. 
Copy is of two types:
    1. Shallow Copy
    2. Deep Copy

Shallow copy: It can be done by "shallow_cpy = dataframe_name.copy(deep=False)". By default parameter 'deep' is True. It creates a new object that shares the refrence to original object. Any changes in shallow copy are reflected in dataframe.

Deep Copy: It can be done by "deep_cpy = dataframe_name.copy(deep=True)". It creates a new object copyin the original object. Any changes in deep copy will not reflect in original object.

```python
samp = data.copy(deep=True)

samp
```

Convert age from months to year

```python
def c_convert(val):
  val_converted = val/12
  return val_converted

samp['Age_Converted']=c_convert(samp['Age'])
samp['Age_Converted']=round(samp['Age_Converted'],1)   ## Round value to one decimal place
```

```python
samp.insert(12,'KM_per_month',0)                       ## Inserting a column named 'KM_per_month' with initial value '0'
```

Single function with parameters age and km returning single output as list of converted age from months to year and km covered per month.

```python
def c_convert(val1,val2):
  val_converted = val1/12
  ratio = val2/val1
  return [val_converted,ratio]

samp['Age_Converted'],samp['KM_per_month']=c_convert(samp['Age'],samp['KM'])
samp['Age_Converted']=round(samp['Age_Converted'],1)
samp['KM_per_month']=round(samp['KM_per_month'],1)       ## Round value to one decimal place
```

## Exploratory Data Analysis

Create Frequency Table

```python
pd.crosstab(samp['FuelType'],columns='count',dropna=True)   ## dropna = True means dropping the rows with NaN values 
## column shows the count of different categories in 'FuelType' column
```

```python
pd.crosstab(samp['Automatic'],columns=samp['FuelType'],dropna=True)  ## Here row and column of table are two different columns of dataframe  
```

Cross_table with multiple columns

```python
pd.crosstab(samp['Automatic'],columns=[samp['FuelType'], samp['Price_Class']],rownames = ['Row'], colnames = ['Column_1', 'Column_2'], dropna=True)  
## Here 'Automatic' column forms the row
## Row is named 'Row' using arguement rownames
## Here 'FuelType' and 'Price_Class' forms the two columns
## Columns are named 'Column_1' and 'Column_2' using arguement colnames
```

Joint probability changes number to proportion. It is probability of two events happening at the same time

```python
pd.crosstab(samp['Automatic'],columns=samp['FuelType'],normalize=True,dropna=True)
```

Marginal Probability = Probability of occurence of single event

```python
pd.crosstab(samp['Automatic'],columns=samp['FuelType'],margins=True,normalize=True,dropna=True)
```

Conditional Probability = Probability of event A given that event B has already occured

```python
pd.crosstab(samp['Automatic'],columns=samp['FuelType'],margins=True,normalize='index',dropna=True)    ## normalize='index' mean normalize over each row -> row sum is one
## Given type of gear box, probability of different fuel_type
```

```python
pd.crosstab(samp['Automatic'],columns=samp['FuelType'],margins=True,normalize='columns',dropna=True)   ## normalize='column' mean normalize over each column -> column sum is one
```
