'read_csv' is the command used to read a comma separated file(csv). Simillarly, in Pandas 'read_excel' is used to read excel files. Pandas reads all the blank values 'NaN'. 

Reading csv files sometimes create an 'Unnnamed: 0' as the first column of the dataframe. So, to avoid this we can specify an argument 'index_col= 0' to read_csv() function, then it reads in the first column as the index.

In the Toypta dataset some values are neither blank nor NaN, they may have some special characters like "??", etc. These values are replaced with NaN using an arguement 'na_values = ['list of unwanted values']' to read_csv() function, now pandas reads these unknown values as NaN.

```python
data = pd.read_csv('Toyota.csv',index_col=0,na_values=['??','???'])
data.dropna(axis=0, inplace=True) #remove all missing values
```

## Scatter Plot
Scater plot is plotted using dots/markers. These markers can vary in size and color. Scatter plot gives us information about the correlation of features i.e. how the two features are related to each other. 

```python
## Here Age and Price show strong negative correlation
plt.scatter(data['Age'],data['Price'],c='red')
plt.title('Age vs Price Plot')
plt.xlabel('Age')
plt.ylabel('Price') 
```

##Histogram (Numerical Data)

Histrogram is plotted for numerical data. It uses bars to represent the frequency (to visualize distribution). We can specify the number of bins, color of bins, color of edge. x-axis represent the continuous variable and y-axis represent the frequency of each bar. 

```python
plt.hist(data['KM'], color='green', edgecolor='white',bins=5)
plt.title('KM Plot')
plt.xlabel('KM')
plt.ylabel('Frequency')
```

##Bar Plot (Categorical Data)

Bar plot is plotted for categorical value. It plots the different categories of any feature. x-axis plots the different categories and y-axis gives the counts of each category. We can decide the color of each category. In bar plot, xticks give labels to category plotted on x-axis.

```python
counts = [979,120,12]
FuelType = ['Petrol','Diesel','CNG']
index = np.arange(len(FuelType))

plt.bar(index, counts, color=['red','cyan','blue'])
plt.title('Bar PLot')
plt.xlabel('Fuel_Type')
plt.ylabel('Count')
plt.xticks(index,FuelType,rotation=90) #rotation rotates the xticks by 90 degree
```

## SEABORN LIBRARY

Seaborn is a visualization library based on the matplotlib. Seaborn is the superset of matplotlib. It can be used for drawing attractive and informative statistical graphics. 
Matplotlib and seaborn both are used for visualization but matplotlib is a library used for basic 2D plots and graphs while seaborn is used for more statistical plots. Matplotlib can be used both in python and matlab. Seaborn support visualizing regression model. 

```python
import seaborn as sns
```

##Scatter Plot

'regplot' stands for regression plot. This method is also used to plot the data model to fit linear regression. 'regplot' plots the scatter plot along with a linear regression model. By default the regression model is shown in graph, if you don't want regression line on the graph then 'fit_reg' arguement can be set 'False'(it is by default true).

```python
sns.set(style="darkgrid")                                               ## a theme is given to the background (here it is 'darkgrid')
# sns.set()                                                             ## By default it will give the 'darkgrid' background
# sns.set_theme(style="darkgrid")                                       ## an alias for 'set' is 'set_theme'
sns.regplot(x=data['Age'],y=data['Price'], fit_reg=False, marker='*')   
```

Plotting scatter plot of two features with respect to third feature. 'lmplot' stands for 'lineplot' which combines 'regplot' and 'facetgrid'. 
An arguement of 'hue' can be added as the third feature with respect to which scatter plot of two features is plotted.

```python
## Scatter plot of 'Age' and 'Price' with respect to 'FuelType'
sns.lmplot(x='Age',y='Price', data = data, hue='FuelType', fit_reg=False, legend='True', palette='husl')  
## 'palette' is used for color palette for coloring points based on data points. 
## Here it is selected as set1 and there are many different color palettes available in seaborn library. For example, pastel, Set1, Set2, flare, etc.
```

##Histogram

Histogram is plotted numerical or continuous variables. It is plotted with default kernel density. 'distplot' stands for Distribution plot. 
By default, it gives the graph of variable on x-axis and kernel density on y-axis, it shows histogram and curve over the bars showing the distribution. 

```python
sns.distplot(data['Age'])
```

If you want to plot the histogram and frequency on y-axis then set the 'kde' arguement as 'False'. Also, 'bins' arguement can be used to specify the number of bins.

```python
sns.distplot(data['Age'], kde=False, bins=5) #without kernel density
## this function may not be available in new versions of python. Also, it may be replaced with 'displot' and 'histplot'.
```

##Bar Plot

##Box and whiskers plot

Box and whiskers plot is one type of graphical representation which shows the five number summary of the data. The smallest value in the data is called the minimum value. The value above it is 25 percentile of data. The value above it is 50 percentile of data, it is also called median. Above it is the 75 percentile of the data. Any value which is less than minimum value and greater than maximum value is called the outlier. 
25 percentile is first quartile and 75 percentile is the third quartile. The difference between two is called interquartile range.

box_whiske plot.jpeg

```python
sns.boxplot(y=data['Price'])      # Median is nearly 10000
# sns.boxplot(x=data['Price'])
```

Box and whisker plot is very useful when we want to check relationship between one numerical and one categorical data.

```python
sns.boxplot(x=data['FuelType'],y=data['Price'])
```

###Grouped Box and Whiskers Plot
Box and whiskers plot of two feattures along with third feature can be formed using a 'hue' parameter.

```python
sns.boxplot(x=data['FuelType'],y=data['Price'], hue=data['Automatic'])
```

##Violin Plot

The violin plots can be inferred as a combination of Box plot at the middle and distribution plots(Kernel Density Estimation) on both side of the data. It gives details like the distribution is mutimodal, its skewness etc.

##Combination of Plots in one window

Combining Histogram & Bar and Whisker plot. This can be done by splitting the window two parts. It is useful if you want to see the distribution of feature and minimum and maximum value at one time.

```python
f,(ax_box, ax_hist) = plt.subplots(2, gridspec_kw={"height_ratios": (.20, .80)})
## subplot function help to plot subplots. Arguement '2' specify the number of rows. Arguenment 'gridspec_kw' tells about the ratio in which these rows is to be divided.

sns.boxplot(x = data['Price'], ax=ax_box)              ##Plots the box plot
sns.distplot(data['Price'], ax=ax_hist, kde=False)     ##Plots the histogram
```

##Joint Plot

It is used to subplot grid for plotting pairwise relationships in a dataset. Joint plot gives three plots one showing the relationship between two variables, plot above and right it tells about the distribution of individual variables.

```python
sns.jointplot(x=data['FuelType'],y=data['Age'], color='pink')     ## The graph in between is scatter plot while the graphs plotted above and right is the histogram.
```

Adding another arguement is possible using 'hue' parameter. If 'hue' parameter is included then the above and right plots are kernel density curves instead of histograms 

```python
sns.jointplot(x=data['Age'],y=data['KM'], hue = data['FuelType']) 
```

Also, you can change the type of bivariate to any other plot using a 'kind' arguement. For example, kind='kde' gives the kernel density curve on all three plots, kind='reg' gives the bivariate plot to be a scatter plot with regression fit line, kind='hist' gives the histogram plot that is bin based visualization, kind='hex' gives the histogram with hexagonal bins, etc.

```python
sns.jointplot(x=data['Age'],y=data['KM'], kind = 'kde', color='green') 
```

##Pairwise plot 

Paiwise plot plots a graph for all pairs of data. Pairs plots can help to quickly explore distributions and relationships in a dataset. It created scatter plot joint relationships and histogram for univariate(diagnol) plots.

```python
sns.pairplot(data)
```

Adding a parameter 'hue' to the pair plot will result in in univariate plots to be layered kernel density curves instead of histogram. Also, you can add an arguement 'kind' to choose the kind of plot of both bivariate and univariate plots. If you want to change the plot type of diagonal plotting then ir can be done using 'diag_kind'. 

```python
sns.pairplot(data, kind='scatter', hue='FuelType')
# sns.pairplot(data, diag_kind='hist', hue='FuelType')
```

```python
sns.pairplot(data, kind='kde', hue='FuelType')
```
