##Dealing with missing value

There are two approaches to handle the missing values. The one approach is to delete the rows which have the missing values. The second approach is to logically. Filling missing value logically varies in case of categorical and numerical data. In case of categorical data it include it include filling missing value with the category having maximum count. In case of numerical data it include filling missing values using central tendencies like mean, mode or median.

It is possible that we sometimes copy of data may be needed because it is possible that we don't want to change the original data. 
Copy is of two types:
    1. Shallow Copy
    2. Deep Copy

Shallow copy: It can be done by "shallow_cpy = dataframe_name.copy(deep=False)". By default parameter 'deep' is True. It creates a new object that shares the refrence to original object. Any changes in shallow copy are reflected in dataframe.

Deep Copy: It can be done by "deep_cpy = dataframe_name.copy(deep=True)". It creates a new object copyin the original object. Any changes in deep copy will not reflect in original object.

##Approach 1: Dropping those rows which has missing data

We can drop those rows which has only one value NaN or deleting those columns which has NaN values. But dropping these rows or columns will result in loss of lage amount of data. Hence, it is good to logically fill the missing data. Also, it is possible to delete those rows which have all features missing.

```python
## Deleting those rows which has one or more NaN values
data.dropna()  ## OR data.dropna(how='any', axis=0)
```

You can see the that initially our data had 4096 rows but after removing rows with NaN values data has only 1096 rows left.

```python
## Delete only those rows which has all values as NaN
data.dropna(how = 'all')
```

```python
## Dropping those columns which has NaN values
data.dropna(axis = 1)
```

##Approach 2: Filling missing values logically

####Fill NaN values for numerical features
For numerical features the NaN values can be filled with mean of the sample or median of the sample.

```python
## Mean and median of Age feature don't have much difference so we can replace NaN values with mean
samp['Age'].fillna(samp['Age'].mean(), inplace=True)    ## Filling NaN values with mean

## Mean and median of KM feature have a huge difference so we can't replace NaN values with mean. 
## The value of mean is very large as compared to median it may be because of any outliers present in KM. 
samp['KM'].fillna(samp['KM'].median(), inplace=True)    ## Filling NaN values with median

## Mean and median of HP feature don't have much difference so we can replace NaN values with mean
samp['HP'].fillna(samp['HP'].mean(), inplace=True)
```

Get the number of null values in each column

```python
samp.isnull().sum()
```

####Fill NaN values for categorical features
For categorical features the NaN values can be filled with the category which has the maximum count.

```python
samp['FuelType'].fillna(samp['FuelType'].value_counts().index[0], inplace=True)
samp['MetColor'].fillna(samp['MetColor'].value_counts().index[0], inplace=True)
```

Get the number of null values in each column

```python
samp.isnull().sum()
```
